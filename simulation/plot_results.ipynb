```python
import matplotlib.pyplot as plt
import json
import os
```

```python
fig, ax = plt.subplots(figsize=(15, 6))

i = 0
for d in ['run_0', 'run_1', 'run_2', 'run_3', 'run_4']:
    results_dir = os.path.join(d, 'results')
    for gt_path in os.listdir(results_dir):
        if not gt_path.startswith("gt_eval"):
            continue
        with open(os.path.join(results_dir, gt_path), 'r') as o:
            gt_eval = json.load(o)
        
        keys = sorted(gt_eval.keys())
        position_error = [gt_eval[k]['p_B_G']['error_norm'] + 0 for k in keys]
        keys = [float(k) for k in keys]
        plt.plot(keys, position_error, label='{}:{}'.format(d, gt_path))
        i += 1

print("Has {} runs.".format(i))
ax.set_ylim(0, 0.01)
#ax.set_yscale('log')
ax.grid(which='major')
ax.grid(which='minor', linestyle='--')
```
